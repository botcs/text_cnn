```python
import re
import codecs


f = codecs.open('./corpus/movie_lines.txt', 'r', encoding='utf-8', errors='ignore')
text = f.readlines()

id2lines = {}
id2names = {}
mID_name = {}
names = set()

lines = []
count = 0
err_line_id = []
for i, line in enumerate(text):
    line = line.replace(']', '').split()
    repaired = ''
    for word in line:
        repaired += word + ' '
        
    repaired = repaired.strip()
    lines.append(repaired)
    
    try:
        lineID, characterID, movieID, character, utt = repaired.split(' +++$+++ ')
    except ValueError:
        err_line_id.append(i)
        #print(repaired, text[i+1])
        #print  '\n\n'
    
    id2lines[lineID] = utt
    id2names[characterID] = character
    if movieID not in mID_name:
        mID_name[movieID] = set()

    
    if len(character) > 1:
        mID_name[movieID].add(character+' ')
        names.add(character+' ')
    

    
```
